```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
timesteps = 30
starting_price = 600
volatility = 0.02
max_sim = 5000
predict=[]
for i in range(max_sim):
    returns=np.random.uniform(1-volatility, 1+volatility, timesteps)
    price=[starting_price]
    for x in returns:
        price.append(price[-1]*x)
    plt.plot(price)
    predict.append(price[-1])
    plt.ylabel('Stock Price')
    plt.xlabel('Time')
    plt.title('Monte Carlo Simultation')
plt.savefig('Paths by MonteCarlo Simultation')
plt.show()
```

```python
plt.hist(predict, bins=100)
plt.ylabel('Occurrence')
plt.xlabel('Price')
plt.title('Distribution')
plt.savefig('Distribution')

plt.show()
```
